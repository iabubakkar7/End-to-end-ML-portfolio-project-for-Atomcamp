#### Importing the necessary libraries

```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

#### Loading the dataframe

```python
data_preprocessed = pd.read_csv("data_preprocessed.csv")
```

```python
data_preprocessed.head(6)
```

```python
targets = data_preprocessed['log_price']
inputs = data_preprocessed.drop(['log_price'],axis=1)
```

```python
data_preprocessed.to_csv('data_preprocessed.csv', index=False)
```

### Scale the data

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(inputs)
```

```python
inputs_scaled = scaler.transform(inputs)
```

### Train Test split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(inputs_scaled, targets, test_size=0.2, random_state=365)
```

### Creating regression

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train,y_train)
```

### Using MLOps to save the model into variable so we can use it anywhere

```python
import joblib
```

```python
joblib.dump(reg, 'my_trained_mode.pkl')
```
